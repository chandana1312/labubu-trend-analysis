## ‚úÖ PHASE 2: Collect Data

```python
from pytrends.request import TrendReq
import pandas as pd

pytrends = TrendReq(hl='en-US', tz=330)
keywords = ['Labubu', 'Bathroom Camping', 'Fake Weddings', 'Yearning']
pytrends.build_payload(keywords, cat=0, timeframe='2022-01-01 2025-07-01', geo='', gprop='')
trend_data = pytrends.interest_over_time()
trend_data.head()
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta

sample_texts = [
    "I just got my first Labubu! üòç", "Labubu-core is real üî•",
    "Honestly Labubu gives me so much serotonin üíñ", "Why is Labubu trending again?",
    "I need more Labubu in my life üê∞‚ú®", "Feeling that #yearning vibe with Labubu rn"
]

# Generate 100 fake tweets from Jan 2022 to July 2025
dates = [datetime(2022, 1, 1) + timedelta(days=random.randint(0, 1275)) for _ in range(100)]
contents = [random.choice(sample_texts) for _ in range(100)]
users = [f"user_{i}" for i in range(100)]

df_labubu = pd.DataFrame({'Date': dates, 'Content': contents, 'User': users})
df_labubu['Date'] = pd.to_datetime(df_labubu['Date'])
df_labubu.head()
```

## ‚úÖ PHASE 3: Clean & Preprocess the Data

```python
df_labubu['Date'] = pd.to_datetime(df_labubu['Date'])
df_labubu['Content'] = df_labubu['Content'].str.lower()
```

```python
from textblob import TextBlob
df_labubu['Sentiment'] = df_labubu['Content'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

## ‚úÖ PHASE 4: Analyze Trends and Gen Z Behavior

```python
import matplotlib.pyplot as plt

trend_data[keywords].plot(figsize=(12,6))
plt.title('üìä Google Search Interest Over Time')
plt.xlabel('Date')
plt.ylabel('Search Interest')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

df_labubu['Month'] = df_labubu['Date'].dt.to_period('M')
monthly_sentiment = df_labubu.groupby('Month')['Sentiment'].mean()
monthly_sentiment.plot(kind='line', figsize=(12,5), title='üí¨ Average Sentiment Over Time (Labubu)')
plt.grid(True)
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(df_labubu['Content'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('‚òÅÔ∏è Labubu Mentions WordCloud')
plt.show()
```

## ‚úÖ PHASE 5: Compare With Gen Z Micro-Trends

```python
import pandas as pd
import random
from datetime import datetime, timedelta
from textblob import TextBlob

# Define mock tweets for each Gen Z trend
mock_texts = {
    "#bathroomcamping": [
        "Bathroom camping is the new self-care ‚õ∫üõÅ", "Just pitched a tent in my tub üòÇ",
        "Another Friday night of bathroom camping ‚ú®", "Honestly, it helps me reset"
    ],
    "#fakeweddings": [
        "Had a fake wedding for content and cried real tears üò≠", "Fake wedding but real drama!",
        "Would you attend a fake wedding for the aesthetic?", "This shoot felt too real üò≥"
    ],
    "#yearning": [
        "Yearning season is back üçÇüíî", "Not sad, just in my yearning era",
        "This playlist is giving peak yearningcore vibes", "I miss people I never met"
    ]
}

queries = ['#bathroomcamping', '#fakeweddings', '#yearning']
dfs = {}

for tag in queries:
    tweets = []
    for i in range(100):  # generate 100 tweets per trend
        date = datetime(2022, 1, 1) + timedelta(days=random.randint(0, 1275))
        content = random.choice(mock_texts[tag])
        user = f"user_{tag[1:]}_{i}"
        tweets.append([date, content, user])

    df = pd.DataFrame(tweets, columns=['Date', 'Content', 'User'])
    df['Date'] = pd.to_datetime(df['Date'])
    df['Content'] = df['Content'].str.lower()
    df['Sentiment'] = df['Content'].apply(lambda x: TextBlob(x).sentiment.polarity)
    df['Month'] = df['Date'].dt.to_period('M')
    dfs[tag] = df

# Preview one dataframe
dfs['#bathroomcamping'].head()
```

```python
plt.figure(figsize=(12,6))
for tag, df in dfs.items():
    monthly_sentiment = df.groupby('Month')['Sentiment'].mean()
    monthly_sentiment.plot(label=tag)
plt.title('üìà Sentiment Comparison Across Gen Z Trends')
plt.xlabel('Month')
plt.ylabel('Average Sentiment')
plt.legend()
plt.grid(True)
plt.show()
```

```python
for tag, df in dfs.items():
    text = " ".join(df['Content'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(12,6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'‚òÅÔ∏è WordCloud for {tag}')
    plt.show()
```

```python
df_labubu.to_csv("labubu_tweets.csv", index=False)
for tag, df in dfs.items():
    df.to_csv(f"{tag.strip('#')}_tweets.csv", index=False)
```
